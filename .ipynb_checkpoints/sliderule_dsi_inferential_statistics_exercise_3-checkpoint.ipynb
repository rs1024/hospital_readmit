# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
# Read in and explore data
data = pd.read_csv('data/cms_hospital_readmissions.csv')
data.head()
```

```python
# Rename cols
data.columns = ['hospital_name', 'provider_number', 'state', 'measure_name', 'number_discharges', 'footnote',
               'excess_readmission_ratio', 'predicted_readmission_rate', 'expected_readmission_rate', 'number_of_readmissions',
               'start_date', 'end_date']
```

```python
# Deal with missing and inconvenient portions of data 
df = data[data.number_discharges != 'Not Available']
df.number_discharges = df.number_discharges.astype(int)
del(df['footnote'])
df = df.dropna()
df = df.sort_values('number_discharges')
```

```python
# Explore distribution of excess readmission ratios grouped by number of charges 
plt.subplot(2, 1, 1)
plt.vlines(df[df.number_discharges<700].excess_readmission_ratio.mean(), 0, 1300, color=sns.xkcd_rgb['sea green'])
plt.hist(df[df.number_discharges<700].excess_readmission_ratio, bins=50)
plt.xlabel('Excess readmission ratio')
plt.ylabel('Count')
plt.title('Distribution of excess readmission ratio for number of discharges < 700')

plt.subplot(2, 1, 2)
plt.vlines(df[df.number_discharges>=700].excess_readmission_ratio.mean(), 0, 260, color=sns.xkcd_rgb['sea green'])
plt.hist(df[df.number_discharges>=700].excess_readmission_ratio, bins=50)
plt.xlabel('Excess readmission ratio')
plt.ylabel('Count')
plt.title('Distribution of excess readmission ratio for number of discharges >= 700')

plt.tight_layout()
None
```

### #1

H<sub>0</sub>: there is no difference between excess readmission ratios for number of discharges < 700 and number of discharges >= 700, i.e. the difference in means = 0  
H<sub>1</sub>: there **is** a difference between the two

```python
mean_less_700 = df[df.number_discharges<700].excess_readmission_ratio.mean()
mean_great_700 = df[df.number_discharges>=700].excess_readmission_ratio.mean()
diff_means = mean_less_700 - mean_great_700

print('Mean excess readmission ratio for hospitals with number of discharges < 700: %f' % (mean_less_700))
print('Mean excess readmission ratio for hospitals with number of discharges >= 700: %f' % (mean_great_700))
print('Difference in sample means: %f' % (diff_means))
```

### #2

```python
def RunModel(df):
    # Resample via bootstrap
    # Randomly select observations from the entire sample as we are assuming no diff between the 2 groups
    # Input a dataframe with n observations and return 2 numpy arrays with n/2 observations
    data1 = np.random.choice(df.excess_readmission_ratio, len(df[df.number_discharges<700]), replace=True)
    data2 = np.random.choice(df.excess_readmission_ratio, len(df[df.number_discharges>=700]), replace=True)
    return data1, data2

def TestStat(data):
    # Returns the test stat which in this case is the difference in means
    data1, data2 = data
    diff_means = data1.mean() - data2.mean()
    return diff_means
```

```python
# Simulate 1000 experiments
test_stats = pd.DataFrame([TestStat(RunModel(df)) for i in range(1000)])
```

```python
plt.vlines(diff_means, 0, 60, color=sns.xkcd_rgb['sea green'])
plt.hist(test_stats, bins=50)
plt.xlabel('Diff in means')
plt.ylabel('Count')
plt.title('Distribution of diff in means under the null')
None
```

```python
z_score = (diff_means - test_stats.mean()) / test_stats.std()
p_value = 1 - stats.norm(0,1).cdf(z_score)

print('z score: %f' % (z_score))
print('p value: %f' % (p_value))
print('\nDiff between sample mean and mean under null: %f' % (diff_means - test_stats.mean()))
```

### #3

p < 0.001. Therefore, we reject the null and conclude that there is likely a difference in excess readmission rates between hospitals with number of discharges < 700 and number of discharges >= 700. Specifically, hospitals with number of discharges < 700 have an average excess readmission ratio that is 0.02 greater than those under 700.

### #4

Statistically, hospitals with discharges < 700 are more likely to have greater excess readmission ratios than those with discharges >= 700 with a p value < 0.001. This finding is very statistically signficant. However in practice, there may be other factors affecting the # of discharges that are not captured in the data and, subsequently, this analysis. Such factors include the size of the hospital, population of the surrounding metropolitan area, # of other hospitals in the vicinity, etc. Thus, there may be confounding variables that we are not capturing which then detracts from the practicality of our recommendations. I would still make recommendations to the client based on my analysis, however I would caution the client to use their judgement when applying to their specific situation.

**Recommendation:** based on my analysis, hospitals with number of discharges < 700 are significantly more likely to have higher excess readmission ratios than those with number of discharges >= 700. I postulate that the size of the hospital is the most significant factor (although there may be other factors not captured), i.e. larger hospitals will have higher numbers of discharges. 

Thus, I believe that smaller hospitals (defined by # discharges < 700) are lacking the resources, personnel, and / or training to properly care for patients, which then results in significantly higher excess readmission ratios. I recommend that smaller hospitals receive additional resources from the state / federal government, consultation as to proper staffing best practices, and government-sanctioned training for all personnel in order to increase the quality of care and reduce overall readmissions.
